```python
import json

# Load the sal.json file
with open('sal.json') as f:
    data = json.load(f)
data
```

```python
from prettytable import PrettyTable
gcc_counts =  {}
for loctions, info in data.items():
    gcc  = info['gcc']
    if gcc[1] != 'r' and gcc in gcc_counts:
        gcc_counts[gcc] += 1
    else:
        gcc_counts[gcc] = 0

table = PrettyTable()
table.field_names = ['Greater Capital City', 'Number of Tweets Made']

for gcc, count in sorted(gcc_counts.items(),key = lambda x:x[1], reverse= True):
    table.add_row([gcc, count])
print(table)

```
